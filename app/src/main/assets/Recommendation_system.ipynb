```python
import pandas as pd
!pip install fuzzywuzzy
!pip install python-Levenshtein
from fuzzywuzzy import process
from sklearn.metrics.pairwise import euclidean_distances

detected_object = "cucumber"

# Function to find the best match for the detected object in the nutritional dataset
def find_best_match(detected_object, itemsp):
    detected_object = detected_object.lower()
    itemsp = [str(item) for item in itemsp]
    if not isinstance(itemsp, list):
     itemsp = list(itemsp) # Convert to lower case to handle case-insensitive matching
    best_match = process.extractOne(detected_object, itemsp, scorer=process.fuzz.partial_ratio)
    if best_match:
        return best_match[0], best_match[1]  # Return both the matched item name and the score
    else:
        return None, None

# Function to recommend top 3 similar items based on nutritional information
def recommend_similar_items(item_name, data, columns=None):
    if item_name in dff['Food and Serving'].values:
        item_nutrition = dff[dff['Food and Serving'] == item_name][columns].values
        other_items = dff[dff['Food and Serving'] != item_name]
        other_nutrition = other_items[columns].values

        # Compute Euclidean distances
        distances = euclidean_distances(item_nutrition, other_nutrition)

        # Get the top 3 smallest distances
        top_indices = distances[0].argsort()[:3]
        recommendations = other_items.iloc[top_indices]

        return recommendations
    else:
        return None

# Call the find_best_match function with the detected object
best_match_name, best_match_score = find_best_match(detected_object, itemsp)

# Print the output to verify the result
print(f"Detected object: {detected_object}")
print(f"Best match: {best_match_name} with score {best_match_score}")

# Specify columns to use for similarity calculation
columns_to_use = ['Calories']

# Recommend the top 3 similar items based on nutritional information
if best_match_name:
    recommendations = recommend_similar_items(best_match_name, n, columns=columns_to_use)
    if recommendations is not None:
        print(f"Top 3 similar items to {best_match_name} based on nutritional information:")
        print(recommendations)
    else:
        print("Item not found in nutritional dataset.")
else:
    print("No match found.")
```
